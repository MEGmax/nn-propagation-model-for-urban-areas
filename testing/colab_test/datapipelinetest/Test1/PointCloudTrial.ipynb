```python
import xml.etree.ElementTree as ET
import open3d as o3d
import numpy as np
import os

# Parse XML
tree = ET.parse("/Users/khushipatel/Desktop/capstone/nn-propagation-model-for-urban-areas/testing/colab_test/datapipelinetest/test1.xml")
root = tree.getroot()

all_points = []

# Loop over all <shape type="ply">
for shape in root.findall(".//shape[@type='ply']"):
    filename_node = shape.find(".//string[@name='filename']")
    if filename_node is not None:
        mesh_path = filename_node.attrib["value"]
        if not os.path.isfile(mesh_path):
            print("File not found:", mesh_path)
            continue

        # Load the PLY mesh
        mesh = o3d.io.read_triangle_mesh(mesh_path)
        if mesh.is_empty():
            print("Warning: empty mesh:", mesh_path)
            continue

        # Sample points from mesh surface
        pcd = mesh.sample_points_uniformly(number_of_points=50000)
        all_points.append(np.asarray(pcd.points))
        print("Loaded and sampled:", mesh_path)

# Merge all points into one point cloud
if all_points:
    merged_points = np.vstack(all_points)

    #pcd final is 
    pcd_final = o3d.geometry.PointCloud()
    pcd_final.points = o3d.utility.Vector3dVector(merged_points)

    # Save to PLY
    o3d.io.write_point_cloud("scene_pointcloud.ply", pcd_final)
    print("Merged point cloud saved as scene_pointcloud.ply")

    # Visualize inline (works in Colab or Jupyter)
    try:
        o3d.visualization.draw([pcd_final])
    except:
        # fallback for older Open3D
        o3d.visualization.draw_geometries([pcd_final])
else:
    print("No points found in the XML meshes!")

```
